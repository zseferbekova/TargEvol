# dN/dS scatter plot 'Mouse vs Chimpanzee'

Import dependencies and data:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("../results/dnds_for_targets.csv", index_col=0)
print(data.shape)
data.head()
```

Filter out inf values:

```python
data_filtered = data.dropna()
print(data_filtered.shape)
data_filtered = data_filtered[(data_filtered["dN/dS with Chimpanzee"] != float("inf")) & (data_filtered["dN/dS with Mouse"] != float("inf"))]
print(data_filtered.shape)
data_filtered.head()
```

```python
sns.set(style="white", font_scale=1.3)

sns.relplot(x="dN/dS with Chimpanzee", 
            y="dN/dS with Mouse", 
            hue="present in DrugBank", 
            height = 5, aspect = 3,
            data=data_filtered) 


plt.title('dN/dS, all genes, all data')

plt.show();
```

```python
sns.set(style="white", font_scale=1.3)

sns.relplot(x="dN/dS with Chimpanzee", 
            y="dN/dS with Mouse", 
            hue="present in DrugBank", 
            height = 5, aspect = 3,
            data=data_filtered[(data_filtered["dN/dS with Chimpanzee"] < 20)]) 


plt.title('dN/dS, all genes, all data')

plt.show();
```

```python
quantile_95_dNdS_Chimp = np.quantile(data_filtered["dN/dS with Chimpanzee"], 0.95)
quantile_95_dNdS_Mouse = np.quantile(data_filtered["dN/dS with Mouse"], 0.95)

plt.figure(figsize=(14, 6))
sns.set(style="white", font_scale=1.3)

sns.relplot(x="dN/dS with Chimpanzee", 
            y="dN/dS with Mouse", 
            hue="present in DrugBank", 
            height = 5, aspect = 3,
            data=data_filtered[(data_filtered["dN/dS with Mouse"] >= quantile_95_dNdS_Mouse) & 
                               (data_filtered["dN/dS with Chimpanzee"] >= quantile_95_dNdS_Chimp)]) 


plt.title('dN/dS, all genes, 95% quantile')

plt.show();
```

```python
quantile_95_dNdS_Chimp = np.quantile(data_filtered["dN/dS with Chimpanzee"], 0.95)
quantile_95_dNdS_Mouse = np.quantile(data_filtered["dN/dS with Mouse"], 0.95)


sns.relplot(x="dN/dS with Chimpanzee", 
            y="dN/dS with Mouse", 
            hue="present in DrugBank", 
            height = 5, aspect = 3,
            data=data_filtered[(data_filtered["dN/dS with Mouse"] >= quantile_95_dNdS_Mouse) & 
                               (data_filtered["dN/dS with Chimpanzee"] >= quantile_95_dNdS_Chimp) & 
                               (data_filtered["dN/dS with Chimpanzee"] < 25)]) 


plt.title('dN/dS, all genes, 95% quantile')

plt.show();
```
